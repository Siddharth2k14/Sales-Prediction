```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
sales_cleaned = sales.drop(columns=['customer name', 'customer e-mail'])
```

```python
X = sales_cleaned.drop(columns=['car purchase amount'])
y = sales_cleaned['car purchase amount']
```

```python
column_transformer = ColumnTransformer(
    transformers = [('onehot', OneHotEncoder(drop="first"), ["country"])],
    remainder="passthrough"
)
```

```python
X_encoded = column_transformer.fit_transform(X)
```

```python
X_encoded_dense = X_encoded.toarray() if hasattr(X_encoded, "toarray") else X_encoded
```

```python
X_encoded_dense
```

```python
feature_names = column_transformer.get_feature_names_out()
```

```python
X_preprocessed = pd.DataFrame(X_encoded_dense, columns=feature_names)
```

```python
X_preprocessed.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_preprocessed, y, test_size=0.2, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_lr = model.predict(X_test)
```

```python
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)
```

```python
print(f"Mean Squared Error (MSE): {mse_lr:.2f}")
print(f"R² Score: {r2_lr:.2f}")
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred_rf = model.predict(X_test)
```

```python
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
```

```python
print(f"Mean Squared Error (MSE): {mse_rf:.2f}")
print(f"R² Score: {r2_rf:.2f}")
```
